### Part 3 - Basic data analysis with Altair (40 points)

##### Question 1 (10 points)

Recreate the figure in Part 3 Question 2 in Homework 1 with Altair. Include necessary title and labels. Here is the question again:
    
    Generate a data visualization to show the relationship between max temperature and weather type.

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import altair as alt
from vega_datasets import data

weather = data.seattle_weather()
weather.head()
```

```python
alt.Chart(weather).mark_boxplot().encode(
    x='weather',
    y='temp_max',
    color='weather'
).properties(
    title='Relationship between Max Temperature and Weather Type'
)
```

Recreate the figure in Part 3 Question 3 in Homework 1 with Altair. Include necessary title and labels. Here is the question again:
    
    Generate a data visualization to show the relationship between max temperature and min temperature, and then show different color for differnet weather type.

```python
alt.Chart(weather).mark_point().encode(
    x='temp_min',
    y='temp_max',
    color='weather'
).properties(
    title='Relationship between Max Temperature and Min Temperature by Weather Type'
)
```

##### Question 3 (10 points)

Create a facet figure in Altair with 2 row and 3 columns. For each panel in the facet figure, it shows a scatter plot between the max and min temperature for one type of weather. Use the wind speed to color the points. Also add interaction to show the max and min temperature and wind speed for each data point. Include necessary title and labels. 

```python
# Create a scatter plot with interaction
scatter = alt.Chart(weather).mark_point().encode(
    x=alt.X('temp_min', title='Min Temperature'),
    y=alt.Y('temp_max', title='Max Temperature'),
    color=alt.Color('wind', title='Wind Speed'),
    tooltip=['temp_min', 'temp_max', 'wind']
).properties(
    width=150,
    height=150
).interactive()

# Facet the scatter plot by weather type
facet = scatter.facet(
    facet=alt.Facet('weather', title='Weather Type'),
    columns=3
).properties(
    title='Scatter Plot of Max and Min Temperature by Weather Type with Wind Speed'
)

facet
```

##### Question 4 (10 points)

Create a compound figure with at least two types of marks. Show the trend of average temperature (the average of min and max temperature) in 4 years. Note, just use the `date` as X attribute, no need to aggragate for years and you will get a wide figure. Include necessary title and labels.

```python
# Calculate the average temperature
weather['avg_temp'] = (weather['temp_max'] + weather['temp_min']) / 2

# Create a line chart for the average temperature trend
line = alt.Chart(weather).mark_line().encode(
    x='date:T',
    y='avg_temp:Q',
    color=alt.value('blue')
).properties(
    title='Trend of Average Temperature Over 4 Years'
)

# Create a point chart for the average temperature trend
points = alt.Chart(weather).mark_point().encode(
    x='date:T',
    y='avg_temp:Q',
    color=alt.value('red')
)

# Combine the line and point charts
compound_chart = line + points

compound_chart.properties(
    width=800,
    height=400
).configure_axis(
    titleFontSize=14,
    labelFontSize=12
).configure_title(
    fontSize=16
)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 5))

# Create a gradient background
gradient = np.linspace(0, 1, 256)
gradient = np.vstack((gradient, gradient))

# Display the gradient background
ax.imshow(gradient, aspect='auto', cmap='gray', extent=[0, 10, 0, 1])

# Add a solid color rectangle in front of the gradient
rect = plt.Rectangle((2, 0.25), 6, 0.5, color='red', alpha=0.7)
ax.add_patch(rect)

# Set axis limits and remove ticks
ax.set_xlim(0, 10)
ax.set_ylim(0, 1)
ax.set_xticks([])
ax.set_yticks([])

plt.show()
```

The above plot shows an example of gradient illusion: the red rectangle appears to change color due to the gradient background. Using numpy I was able to produce a gradient background, on top of which I placed a static rectangle.
